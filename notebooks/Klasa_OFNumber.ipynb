__Funkcje matematyczne:__

```python
from ofnumber import fexp, fabs, flog, fpower, fmax, fmin
a = OFNumber.init_trapezoid_x0x1(-1, 2, 5, 3, dim=6)
```

```python
b = fabs(a)
fig, ax = plt.subplots()
a.plot_ofn(ax, plot_as='classic')
b.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```

```python
b = fexp(a)
fig, ax = plt.subplots()
a.plot_ofn(ax, plot_as='classic')
b.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```

```python
a = OFNumber.init_trapezoid_x0x1(1, 2, 5, 3, dim=6)
b = flog(a)
fig, ax = plt.subplots()
a.plot_ofn(ax, plot_as='classic')
b.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```

```python
a = OFNumber.init_trapezoid_x0x1(1, 2, 5, 3, dim=6)
b = fpower(2, a)
fig, ax = plt.subplots()
a.plot_ofn(ax, plot_as='classic')
b.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```

```python
a = OFNumber.init_trapezoid_x0x1(1, 2, 5, 3, dim=6)
b = fmin(a, 2)
fig, ax = plt.subplots()
a.plot_ofn(ax, plot_as='classic')
b.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```
