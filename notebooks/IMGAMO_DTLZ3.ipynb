```python
import numpy as np
import imgamo
import problems
from problems import Kursawe, Problem, DTLZ3, DTLZ2
from players import ClonalSelection, SimulatedAnnealing, SimpleGeneticAlg, ScipyMinimize
from operators import hiper_mutate
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
nobjs = 3
nvars = 7
#problem = DTLZ2(nobjs=nobjs, nvars=9, need_repair=False)
problem = DTLZ3(nobjs=nobjs, nvars=nvars, need_repair=False)
```

```python
players = [ClonalSelection(i, nclone=15, supp_level=0, mutate=hiper_mutate, mutate_args=(0.45, 0.9, 0.0001))
           for i in range(nobjs)]
#players = [ScipyMinimize(i, niter=1) for i in range(nobjs)]
```

```python
options = imgamo.Options(population_size=25, max_evaluations=-1, max_iterations=1000, exchange_iter=3, change_iter=3,
                         front_max=200, seed=None, verbose=100)
```

```python
solver = imgamo.IMGAMO(problem, players, options)
```

```python
solver.run_algorithm()
```

```python
solver.result.summary()
```

```python
%matplotlib notebook
ax = plt.figure().add_subplot(111, projection='3d')
solver.result.plot_3d(ax, 0, 1, 2)
plt.show()
```
