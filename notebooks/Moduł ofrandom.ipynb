```python
import sys
sys.path.append('../pyorderedfuzzy')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import ofnumber as ofn
import ofmodels as ofm
import ofrandom as ofr
```

```python
mu = ofn.OFNumber.init_trapezoid_x0x1(1, 2, 5, 3)
sig2 = ofn.OFNumber.init_trapezoid_x0x1(0.0, 0.02, 0.0, 0.02) #ofn.OFNumber.init_from_scalar(0.002)
s2 = 1.
p = 0.5
```

```python
ran = ofr.ofnormal(mu, sig2, s2, p)
```

```python
fig, ax = plt.subplots()
ran.plot_ofn(ax, plot_as='classic')
plt.show()
```

```python
o = ofr.ofnormal_sample(10000, mu, sig2, s2, p)
```

```python
fig, ax = plt.subplots()
o.plot_ofseries(ax, e=10)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
o.plot_histogram(ax1, ax2, 0.5, bins=20)
plt.show()
```

```python
mu_est = ofr.ofnormal_mu_est(o)
fig, ax = plt.subplots()
mu.plot_ofn(ax, plot_as='classic')
mu_est.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```

```python
sig2_est = ofr.ofnormal_sig2_est(o)
fig, ax = plt.subplots()
sig2.plot_ofn(ax, plot_as='classic')
sig2_est.plot_ofn(ax, plot_as='classic', kwargs_f={'c': 'r'}, kwargs_g={'c': 'r'})
plt.show()
```
