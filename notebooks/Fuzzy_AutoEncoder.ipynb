```python
import numpy as np
import pandas as pd
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'../data/WIG20.prn')
df.head()
```

```python
df['DateTime'] = df['Date'].astype(str)+df['Time'].astype(str)
df.head()
```

```python
def conv_time(x):
    if len(x)==12:
        return pd.datetime(int(x[:4]), int(x[4:6]), int(x[6:8]), int(x[8:10]), int(x[10:]))
    elif len(x)==11:
        return pd.datetime(int(x[:4]), int(x[4:6]), int(x[6:8]), int(x[8:9]), int(x[9:]))
    else:
        raise ValueError('len(x)={}'.format(len(x)))
        

df['DateTime'] = df['DateTime'].astype(str).apply(conv_time)
```
